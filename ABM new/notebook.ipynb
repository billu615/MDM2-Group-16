```python
import seaborn as sns
from pollinator_model import PollinatorModel
```

```python
print('processing')
model = PollinatorModel(bee_type='honeybee', sensitivity='moderate', width=500, height=500, num_pollinators=500)
for _ in range(1000):
    model.step()
data = model.datacollector.get_model_vars_dataframe()
print(data)
g = sns.lineplot(data=data['Total Pollinators'])
g.set(title='Honeybee Population Decay', xlabel='timesteps')
```
